```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
data_dir = Path('C:\\Users\\Yarden Cohen\\repos\\tweetynet\\results\\Canaries')
data_csv_5ms = sorted(data_dir.glob('*5ms*csv'))
data_csv_10ms = sorted(data_dir.glob('*10ms*csv'))
```

```python
with plt.xkcd():
    fig = plt.figure(figsize=(12,8))
    cnt=1
    for fname in data_csv_5ms:
        df = pd.read_csv(fname)
        ax=fig.add_subplot(3,4,(cnt-1)*4+1)
        ax.boxplot(np.transpose(1-df.iloc[:,:4]),vert=True,sym='')
        #ax.set_xticks([1,2,3,4])
        if cnt==3:
            ax.set_xticklabels(['naive','maj','min_dur','main_dur+maj'],rotation=45)
        else:
            ax.set_xticks([])
        ax.set_title(df['bird_id'][0]+'_5ms')
        ax.set_ylabel('Frame error rate')
        ax.set_ylim([0,0.06])
        
        ax=fig.add_subplot(3,4,(cnt-1)*4+2)
        ax.boxplot(np.transpose(df.iloc[:,8:12]),vert=True,sym='')
        ax.set_title(df['bird_id'][0]+'_5ms')
        ax.set_ylabel('Segment error rate')
        ax.set_ylim([0,0.2])
        if cnt==3:
            ax.set_xticklabels(['naive','maj','min_dur','main_dur+maj'],rotation=45)
        else:
            ax.set_xticks([])
        cnt+=1
        
cnt=1
with plt.xkcd():
   
    for fname in data_csv_10ms:
        df = pd.read_csv(fname)
        ax=fig.add_subplot(3,4,(cnt-1)*4+3)
        ax.boxplot(np.transpose(1-df.iloc[:,:4]),vert=True,sym='')
        #ax.set_xticks([1,2,3,4])
        if cnt==3:
            ax.set_xticklabels(['naive','maj','min_dur','main_dur+maj'],rotation=45)
        else:
            ax.set_xticks([])
        ax.set_title(df['bird_id'][0]+'_10ms')
        ax.set_ylabel('Frame error rate')
        ax.set_ylim([0,0.06])
        
        ax=fig.add_subplot(3,4,(cnt-1)*4+4)
        ax.boxplot(np.transpose(df.iloc[:,8:12]),vert=True,sym='')
        ax.set_title(df['bird_id'][0]+'_10ms')
        ax.set_ylabel('Segment error rate')
        ax.set_ylim([0,0.2])
        if cnt==3:
            ax.set_xticklabels(['naive','maj','min_dur','main_dur+maj'],rotation=45)
        else:
            ax.set_xticks([])    
            
        cnt +=1
plt.tight_layout()
```
