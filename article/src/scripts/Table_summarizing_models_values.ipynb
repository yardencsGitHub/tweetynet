```python
# Canary - training duration 240 - separate birds
df_lc_240 = copy.deepcopy(df_lc[df_lc.train_set_dur==240])
birds = np.unique(df_lc.animal_id)
for cleanup in cleanups:
    print('cleanup - ',cleanup)
    print('Mean frame error:',1-np.mean([np.mean(df_lc_240[(df_lc_240.cleanup == cleanup) & (df_lc_240.animal_id==bird)].avg_acc) for bird in birds]))
    print('Std frame error:',np.std([np.mean(df_lc_240[(df_lc_240.cleanup == cleanup) & (df_lc_240.animal_id==bird)].avg_acc) for bird in birds]))
    print('Mean seg. error:',np.mean([np.mean(df_lc_240[(df_lc_240.cleanup == cleanup) & (df_lc_240.animal_id==bird)].avg_segment_error_rate) for bird in birds]))
    print('Std seg. error:',np.std([np.mean(df_lc_240[(df_lc_240.cleanup == cleanup) & (df_lc_240.animal_id==bird)].avg_segment_error_rate) for bird in birds]))
    print('Mean in boundary:',np.mean(df_lc_240[df_lc_240.cleanup == cleanup].avg_pct_boundary_err))
    print('Std in boundary:',np.std(df_lc_240[df_lc_240.cleanup == cleanup].avg_pct_boundary_err))
```
