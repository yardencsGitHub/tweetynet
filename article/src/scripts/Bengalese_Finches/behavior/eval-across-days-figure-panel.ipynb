```python
import numpy as np
import pandas as pd
import seaborn as sns

import pyprojroot
```

```python
CSV_FNAME = 'eval-across-days.csv'
RESULTS_ROOT = pyprojroot.here()  / 'results' / 'Bengalese_Finches' / 'behavior'

eval_across_days_csv = RESULTS_ROOT / CSV_FNAME
df = pd.read_csv(eval_across_days_csv)
```

```python
df_minsegdur_majvote = df[df.cleanup == 'min_segment_dur_majority_vote']
```

```python
sns.set()
```

```python
FIG_ROOT = pyprojroot.here() / 'doc' / 'article' / 'figures' / 'mainfig_bf_behavior'
FIG_STEM = 'segment_error_across_days'
```

```python
g = sns.relplot(data=df_minsegdur_majvote, 
                x='day_int', y='avg_segment_error_rate',
                kind='line', 
                hue='animal_id', 
                ci='sd',
                aspect=1.5,
               )
g.fig.set_dpi(300)
g.set(xticks=[1, 2, 3, 4])
g.axes[0][0].set_ylim([0., 0.3])
g.axes[0][0].set_ylabel('segment error rate')
g.axes[0][0].set_xlabel('day');

for ext in ('eps', 'pdf', 'svg'):
    g.fig.savefig(FIG_ROOT / f'{FIG_STEM}.{ext}')
```
