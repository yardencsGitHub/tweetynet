```python
import pandas as pd
import pyprojroot
import seaborn as sns
```

```python
STATS_CSV_PATH = pyprojroot.here() / 'results' / 'tweetynet_vs_svm' / 'stats.csv'
```

```python
df = pd.read_csv(STATS_CSV_PATH)
```

```python
g = sns.FacetGrid(df, col="species", row="segmentation",
                  palette="Set3", height=4, aspect=2)
g.map(sns.barplot, 'train_dur', 'pval_wilcoxon', order=sorted(df.train_dur.unique()))

for ax_ in g.axes.flatten():
    ax_.axhline(y=0.05, linestyle='--')
```

```python
g = sns.FacetGrid(df, col="species", row="segmentation",
                  palette="Set3", height=4, aspect=2)
g.map(sns.barplot, 'train_dur', 'pval_levene', order=sorted(df.train_dur.unique()))

for ax_ in g.axes.flatten():
    ax_.axhline(y=0.05, linestyle='--')
```
